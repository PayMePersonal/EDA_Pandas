```python
import pandas as pd
import glob
import os



# Set options to display all rows, columns, and full content
pd.set_option('display.max_rows', None)     # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_colwidth', None) # Show full content of cells

# Define the path to the folder containing the Excel files
# Use a pattern to match all .xlsx files (adjust for .xls if needed)
path = 'datasetfiles/*.xlsx' 

# Get a list of all file paths that match the pattern
all_files = glob.glob(path)

# Create an empty list to store individual DataFrames
list_of_dfs = []

# Loop through the list of file paths
for filename in all_files:
    # Read the Excel file into a DataFrame
    # pd.read_excel() reads only the first sheet by default,
    # or you can specify 'sheet_name=None' to get a dictionary of all sheets
    df = pd.read_excel(filename, sheet_name=None) 
    
    # If the file has multiple sheets and you want all data in one DF, 
    # you will need another inner loop or use pd.concat on the dictionary values
    # For this example, assuming each file has one relevant sheet, or all sheets 
    # have the same structure and you combine them first:
    
    # A simple way to handle potential multiple sheets in one file is to process the dictionary:
    if isinstance(df, dict):
        df = pd.concat(df.values(), ignore_index=True)

    list_of_dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
full_df = pd.concat(list_of_dfs, ignore_index=True)

# Display the final DataFrame
print(full_df)
```
