```python
import pandas as pd

# Create DataFrame 1 (df1)
data1 = {'ID': [1, 2, 3, 4],
         'Name': ['Alice', 'Bob', 'Charlie', 'David'],
         'Age': [25, 30, 35, 40]}
df1 = pd.DataFrame(data1)

# Create DataFrame 2 (df2)
data2 = {'ID': [1, 2, 3, 5],
         'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
         'Zip': [10001, 90001, 60601, 77001]}
df2 = pd.DataFrame(data2)

# Perform an inner merge on the 'ID' column
# An inner merge keeps only the rows with matching IDs in both DataFrames
merged_df = pd.merge(df1, df2, on='ID', how='inner')

print(merged_df)
```
